<font color="##0000dd" size="3">**Workspace**: no missing values, but APP_ID is string</font>

```python
# show the row where APP_ID is string
wspace[wspace['APP_ID'].apply(lambda x: isinstance(x, str))]
```

```python
# delete the row where APP_ID is string
wspace = wspace[wspace['APP_ID'].apply(lambda x: isinstance(x, int))]
wspace['APP_ID'] = wspace['APP_ID'].astype(int)
wspace.info()
```

```python
print('Number of users who interacted with apps is:', log['USER_PID'].nunique())
print('Number of apps being used is:', log['APP_ID'].nunique())
log_user = log.USER_PID.unique().tolist()
log_app = log.APP_ID.unique().tolist()
```

# Combine Dataframes

```python
# this shows how user interact with apps
user_log = pd.merge(user, log, on='USER_PID')
user_log.head()
```

```python
# user & mean_usage_time matrix
user_log.pivot_table(index='USER_PID',columns='APP_ID',values='USAGE_TIME_MINS').head()
```

```python
# this shows how user rate apps
user_rating = pd.merge(user, rating, on='USER_PID')
user_rating.head()
```

```python
# user & rating matrix
user_rating.pivot_table(index='USER_PID',columns='APP_ID',values='OVERALL_RATING').head()
```

```python
# this shows app category and user's ways to organize apps
wspace_app = pd.merge(app, wspace, on='APP_ID', how='left')
wspace_app.head()
```
